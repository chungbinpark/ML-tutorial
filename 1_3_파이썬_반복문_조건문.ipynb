***삼각형 넓이 함수 만들기***

```python
def area(a,b,c):
    s = (a+b+c)/2
    result = (s*(s-a)*(s-b)*(s-c))**0.5
    return result
```

```python
area(2,3**0.5,1)
```

```python
area(4, 2*3**0.5,2)
```

# **원주율 구하기**

pi2.png



1.   rand() 함수 사용
2.   입력값을 시행횟수, 출력값을 계산된 $\pi$ 값으로 하는 함수 만들기
3. for문, if문 사용



```python
import numpy as np

def cal_pi(number):
    num_in = 0
    for i in range(0,number):
        x = np.random.rand()
        y = np.random.rand()
        r = (x**2+y**2)**0.5
        if r <= 1.0:
            num_in = num_in + 1
    expected_pi = float(num_in/number)*4.0
    return expected_pi
```

```python
cal_pi(10)
```

```python
cal_pi(100)
```

```python
cal_pi(1000)
```

```python
for i in range(0,8):
  n = 10**(i+1) # 0 ~ 4
  pi = cal_pi(n)
  print ("시행횟수 = 10^",(i+1),":",pi)
```

```python
num_in = 0
num_out = 0
for i in range(0,10):
    x = np.random.rand()
    y = np.random.rand()
    r = (x**2+y**2)**0.5
    if r <= 1.0:
        num_in = num_in + 1
    else:
        num_out = num_out + 1
    print ("in =",num_in, "out =",num_out)
```
