## **예제. 삼각형 넓이 구하기**
삼각형의 세변의 길이(a, b, c)를 통해 넓이를 구하는 코드를 작성한다.
*   $area = \sqrt{s(s-a)(s-b)(s-c)}$
*   $s=\frac{a+b+c}{2}$
*   $a = 2, b = \sqrt{3}, c= 1$





```python
a = 2.0
b = 3.0**0.5
c = 1.0
s = (a+b+c)/2.0
area = (s*(s-a)*(s-b)*(s-c))**0.5
area
```
