**(3) 배열의 각 원소 (인덱스)**

*   첫번째 학생 점수: 92
*   두번째 학생 점수: 83
*   세번째 학생 점수: 87
*   네번째 학생 점수: 97

```python
score =[92,83,87,97,81]
score
```

```python
score[0]
```

다섯번째 학생의 점수는?

```python
score[4]
```

**(4) 슬라이싱: 배열의 원소 중 일정한 부분만 가져오기**

list[a:b] : 인덱스 a 번부터 (b-1)번까지 가져오기

```python
list = [1,3,5,7,9,11,13,15]
```

인덱스 2번부터 5번까지 가져오기

```python
list[2:6]
```

인덱스 2번부터 마지막까지 가져오기

```python
list[2:]
```

인덱스 0번부터 4번까지 가져오기

```python
list[:5]
```

*   넘파이: 배열, 행렬 계산에 필요한 함수들을 모아 놓은 파이썬 라이브러리
*   numpy 데이터: 빠른 수치 계산용 데이터





```python
import numpy # numpy 라이브러리 불러오기
```

```python
import numpy as np # numpy 라이버러리를 가져오며, 함수 호출 시 np 라는 이름을 쓰겠다.
```

```python
darray = np.array([[1,2,3,4],[5,6,7,8]])
```

**(4) 기타 넘파이 배열 만들기**

```python
zero = np.zeros((2,5)) # 2X5 2차원 넘파이 배열을 만들고 그 성분은 모두 0으로 한다.
zero
```

```python
one = np.ones((3,6)) # 3X6 2차원 넘파이 배열을 만들고 그 성분은 모두 1로 한다.
one
```
