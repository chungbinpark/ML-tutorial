예제 1. $a_{n+1} = 2a_{n}+3$이고 $a_0 = 8$ 일 때, $a_5$를 구하세요.

```python
a = 8
for i in range(0,6):
    a = 2*a+3
    print (i+1,a)
```

예제 2. Newton method를 활용하여 $x^2-2x-3 = 0$ 의 해 구하기

```python
x = -10
step = 10
for i in range(0,step):
    m = 2*x-2
    y = x**2-2*x-3
    x = x-y/m
    print (x)
```
